```python
def create_barplot(categories, tfidf_values, word2vec_values, title, save_filename=None):
   
    bar_width = 0.35
    # bar position
    r1 = range(len(categories))
    r2 = [x + bar_width for x in r1]

    # barplot creation
    bars1 = plt.bar(r1, tfidf_values, color='blue', width=bar_width, edgecolor='grey', label='TF IDF')
    bars2 = plt.bar(r2, word2vec_values, color='darkorange', width=bar_width, edgecolor='grey', label='WORD2VEC')

   # x label
    plt.xticks([r + bar_width/2 for r in range(len(categories))], categories)
    plt.ylim(0.25, 1.0)
    plt.legend(fontsize='x-small', loc='upper left', bbox_to_anchor=(1, 1))

    # percentage labels on the bars
    for bar, value in zip(bars1, tfidf_values):
        plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 0.01, f'{value*100:.2f}%', ha='center', va='bottom', color='black')

    for bar, value in zip(bars2, word2vec_values):
        plt.text(bar.get_x() + bar.get_width() / 2 + 0.1, bar.get_height() + 0.01, f'{value*100:.2f}%', ha='center', va='bottom', color='black')

    # figure margins
    plt.subplots_adjust(top=1, bottom=0.025)

    # title settings
    plt.title(title, fontweight='bold', fontsize='large', y=1.065)
    if save_filename:
        plt.savefig(fname=save_filename, format='jpeg', bbox_inches='tight')
    plt.show()

categories = ['Accuracy', 'Mean Precision', 'Mean Recall', 'F1 Score']
#Naive-Bayes
tfidf_values_1 = [0.92, 0.915, 0.795, 0.851] 
word2vec_values_1 = [0.7, 0.3775, 0.455, 0.413]
#KNN
tfidf_values_2 = [0.93, 0.89, 0.86, 0.875]
word2vec_values_2 = [0.92, 0.875, 0.845, 0.859]
#SVM
tfidf_values_3 = [0.95, 0.93, 0.86, 0.893]
word2vec_values_3 = [0.92, 0.8625, 0.8175, 0.839]
```

```python
# Call the function 
create_barplot(categories, tfidf_values_1, word2vec_values_1, 'Summary of Multinomial Naive Bayes (MNB) classification result diagram',
               save_filename=f'{IMAGES_PATH}/MNB_barplot.png')
create_barplot(categories, tfidf_values_2, word2vec_values_2, 'Summary of K-Nearest Neighbours (KNN) classification result diagram, k=3',
               save_filename=f'{IMAGES_PATH}/KNN3_barplot.png')
create_barplot(categories, tfidf_values_3, word2vec_values_3, 'Summary of Support Vector Machine (SVM) classification result diagram',
               save_filename=f'{IMAGES_PATH}/SVM_barplot.png')
```

## COMPARISONS OF GOODNESS MEASURES

```python
def create_comparison_barplot(models, tfidf_values, word2vec_values, y_label, title, save_filename=None):
    bar_width = 0.35

    r1 = np.arange(len(models))
    r2 = [x + bar_width for x in r1]

    bars1 = plt.bar(r1, tfidf_values, color='blue', width=bar_width, edgecolor='grey', label='TF IDF')
    bars2 = plt.bar(r2, word2vec_values, color='darkorange', width=bar_width, edgecolor='grey', label='WORD2VEC')

    plt.xlabel('Classifiers', fontweight='bold')
    plt.xticks([r + bar_width/2 for r in range(len(models))], models)

    plt.ylabel(y_label, fontweight='bold')
    plt.ylim(0, 1.0)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.title(title)

    for bar, value in zip(bars1, tfidf_values):
        plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 0.01, f'{value*100:.2f}%', ha='center', va='bottom', color='black')

    for bar, value in zip(bars2, word2vec_values):
        plt.text(bar.get_x() + bar.get_width() / 2 + 0.1, bar.get_height() + 0.01, f'{value*100:.2f}%', ha='center', va='bottom', color='black')

    plt.subplots_adjust(top=1, bottom=0.025)

    if save_filename:
        plt.savefig(fname=save_filename, format='jpeg', bbox_inches='tight')
    plt.show()
```

```python
models = ['Naive-Bayes', 'KNN', 'SVM']
# FUNCTION CALLS
tfidf_recall_1 = [0.795, 0.86, 0.86]
word2vec_recall_1 = [0.455, 0.845, 0.8175]
create_comparison_barplot(models, tfidf_recall_1, word2vec_recall_1, 
                          'Recall', 'Comparison of Mean RECALL diagram', save_filename = f'{IMAGES_PATH}/RECALL_comparison.png')


tfidf_accuracy_2 = [0.92, 0.93, 0.95]
word2vec_accuracy_2 = [0.68, 0.92, 0.92]
create_comparison_barplot(models, tfidf_accuracy_2, word2vec_accuracy_2, 
                          'Accuracy', 'Comparison of ACCURACY diagram', save_filename = f'{IMAGES_PATH}/ACCURACY_comparison.png')


tfidf_f1_score_3 = [0.850789474, 0.874742857, 0.893631285]
word2vec_f1_score_3 = [0.412642643, 0.859738372, 0.839397321]
create_comparison_barplot(models, tfidf_f1_score_3, word2vec_f1_score_3, 
                          'F1 Score', 'Comparison of F1 SCORE diagram', save_filename = f'{IMAGES_PATH}/F1_SCORE_comparison.png')


tfidf_precision_4 = [0.915, 0.89, 0.93]
word2vec_precision_4 = [0.3775, 0.875, 0.8625]
create_comparison_barplot(models, tfidf_precision_4, word2vec_precision_4,
                          'Precision', 'Comparison of Mean PRECISION diagram', save_filename = f'{IMAGES_PATH}/PRECISION_comparison.png')
```
